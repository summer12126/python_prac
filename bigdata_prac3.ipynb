```python
import seaborn as sns
import pandas as pd
df=sns.load_dataset('planets')
df
```

```python
num_index = int(len(df.index)*0.7)
df_selected = df[:num_index] 
```

```python
df_data= df_selected.copy()
df_data['distance'] = df_data['distance'].fillna(df_data['distance'].median())
df_data['distance']

```

```python
import numpy as np

print(f"origin: {np.std(df_selected.distance)}, preprocessed: {np.std(df_data.distance)}")
```
