```python
import seaborn as sns
import pandas as pd
df=sns.load_dataset('car_crashes')
print(df)
```

```python
df_sorted_alcohol = df.sort_values('alcohol' ,ascending=False)
```

```python
ten = df_sorted_alcohol.iloc[9,2]
```

```python
df_sorted_alcohol.iloc[:9,2] = ten
df_result = df_sorted_alcohol[df_sorted_alcohol.speeding >=7]
alcohol= df_result["alcohol"].mean() 
```

```python
print(f"Final result: {alcohol}")
```
